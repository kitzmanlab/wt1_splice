```python
import splanl.junction_scorer as jn
import splanl.merge_bcs as mbcs
import splanl.coords as cds
import splanl.plots as sp
import splanl.score_motifs as sm
import splanl.inspect_variants as iv
import splanl.post_processing as pp
import splanl.custom_splai_scores as css
```

```python
chr11 = '/nfs/kitzman3/lab_common/refs/human/ucsc_hg19/bychrom/chr11.fa'
chr11_seq = pp.get_refseq( chr11 )[ 0 ]
```

```python
data_in = '/nfs/kitzman2/smithcat/proj/wt1_2022/ex9_data/'
```

```python
byvartbl_wide = pd.read_table( data_in + 'wt1_ex9_by_var_effects_snvs_wide-2022-0721.txt' )
```

```python
lit = pd.read_table( data_in + 'wt1_ex9_variant_lit.2022-0721.txt' )
```

```python
lit
```

```python
lit = lit.rename( columns = { 'cDNA ref': 'ref_c',
                              'cDNA alt': 'alt_c',
                              'cDNA_pos': 'hgvs_pos' } )
```

```python
lit[ 'ref' ] = [ css.rev_complement( r ) for r in lit.ref_c ]
```

```python
lit[ 'alt' ] = [ css.rev_complement( a ) for a in lit.alt_c ]
```

```python
#ran this as a left merge bc one of the lit variants is too far into US intron (c.1355-87)
idx_cols = [ 'hgvs_pos', 'hg19_pos', 'ref', 'alt', 'ref_c', 'alt_c' ]

byvartbl_wide = byvartbl_wide.set_index( idx_cols ).merge( lit.set_index( idx_cols ),
                                                           how = 'left',
                                                           left_index = True,
                                                           right_index = True ).reset_index()
```

```python
byvartbl_wide[ 'clinvar' ] = byvartbl_wide[ 'ClinVar interp' ].notnull()
```

```python
byvartbl_wide[ 'acmg' ] = byvartbl_wide[ 'Interpretation_ACMG_Varsome' ].notnull()
```

```python
( ( byvartbl_wide.clinvar ) | ( byvartbl_wide.acmg ) ).sum()
```

```python
sns.swarmplot( x = 'ClinVar interp', y = 'KTS_+_ratio_log10', data = byvartbl_wide, s = 4 )
plt.xticks( rotation = 90 )
sns.despine()
```

```python
byvartbl_wide[ 'lit_interp' ] = [ 'LPP' if i == 'pathogenic' or i == 'likely pathogenic' \
                                  else 'VUS' if i == 'VUS' \
                                  else 'Conflicting' if i == 'conflicting interpretations' \
                                  else 'LBB' if i == 'benign' or i == 'likely benign' \
                                  else np.nan for i in byvartbl_wide[ 'ClinVar interp' ] ]
```

```python
byvartbl_wide.loc[ ( byvartbl_wide.Interpretation_ACMG_Varsome == 'Pathogenic' ) & ( byvartbl_wide.lit_interp.isnull() ), 'lit_interp' ] = 'Lit_LPP'
```

```python
sns.swarmplot( x = 'lit_interp', y = 'KTS_+_ratio_log10', data = byvartbl_wide, s = 4 )
plt.xticks( rotation = 90 )
sns.despine()
```

```python
byvartbl_wide[ 'KTS_+_ratio_log2' ] = np.log2( byvartbl_wide[ 'wmean_KTS+'] / byvartbl_wide[ 'wmean_KTS-' ] )
```

```python
sns.swarmplot( x = 'lit_interp', y = 'KTS_+_ratio_log2', data = byvartbl_wide, s = 4 )
plt.xticks( rotation = 90 )
sns.despine()
```

```python
np.random.seed( 721 )

fig,ax = plt.subplots( 2, 1,
                       sharex = True,
                       figsize = ( 10, 10 ) )

markers = [ '+', '*', '.', 'D', 'v' ]

tbv = byvartbl_wide.copy()

tbv = tbv.loc[ tbv.lit_interp.notnull() ].copy()

for i,interp in enumerate( tbv.lit_interp.unique() ):
    
    n_interp = ( tbv.lit_interp == interp ).sum()
    
    if n_interp > 1:
    
        offsets = np.random.normal( 0, .2, n_interp )
    
    else:
        
        offsets = [ 0 ]
    
    ax[ 0 ].scatter( [ i + off for off in offsets ],
                     tbv.loc[ tbv.lit_interp == interp ][ 'KTS_+_ratio_log2' ],
                     marker = markers[ i ],
                     s = 200 )
    
    ax[ 1 ].scatter( [ i + off for off in offsets ],
                     tbv.loc[ tbv.lit_interp == interp ].wmean_SKIP,
                     marker = markers[ i ],
                     s = 200 )
    
ax[ 0 ].set_ylabel( 'KTS+ ratio log2' )
ax[ 1 ].set_ylabel( 'SKIP PSI' )

ax[ 1 ].set_ylim( ( 0, 1 ) )

ax[ 1 ].set_xlabel( 'Clinical Interpretation' )

ax[ 1 ].set_xticks( list( range( len( tbv.lit_interp.unique() ) ) ) )
ax[ 1 ].set_xticklabels( tbv.lit_interp.unique() )

plt.show()
```
